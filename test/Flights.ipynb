```python
!cdn_resources='in_line'

g= nx.Graph()
Colors={"A":"#6676F5","B":"#F5F46D","Ch":"#64D5D9","Ci":"#E498F5","Co":"#D16F68"}
fig="diamond"
g.add_node('A', label='AMEX MR',color=Colors["A"],size=15,shape=fig)
g.add_node("B",label="Bilt",color=Colors["B"],size=15,shape=fig)
g.add_node('Ci', label='Citi',color=Colors["Ci"],size=15,shape=fig)
g.add_node('Ch', label='Chase',color=Colors["Ch"],size=15,shape=fig)
g.add_node("Co",label="Capital One",color=Colors["Co"],size=15,shape=fig)


g.add_nodes_from(star,group="star")
g.add_nodes_from(one,group="one")
g.add_nodes_from(sky,group="sky")


for i in range(len(star)-1):
  for j in range(i+1,len(star)):
    g.add_edge(star[i],star[j],color="lightgray")
for i in range(len(one)-1):
  for j in range(i+1,len(one)):
    g.add_edge(one[i],one[j],color="lightgray")

for i in range(len(sky)-1):
  for j in range(i+1,len(sky)):
    g.add_edge(sky[i],sky[j],color="lightgray")

#Indivual Airlines
for i in range(len(AS)):
  g.add_edge("Alaska",AS[i],color="lightgray")
for i in range(len(VA)):
  g.add_edge("Virgin Atlantic",VA[i],color="lightgray")
for i in range(len(VAs)):
  g.add_edge("Virgin Australia",VAs[i],color="lightgray")
for i in range(len(JB)):
  g.add_edge("JetBlue",JB[i],color="lightgray")
for i in range(len(JAL)):
  g.add_edge("JAL",JAL[i],color="lightgray")


for i in AMEX:
  g.add_edge("A",i,color=Colors["A"])
for i in Citi:
  g.add_edge("Ci",i,color=Colors["Ci"])
for i in Bilt:
  g.add_edge("B",i,color=Colors["B"])
for i in Chase:
  g.add_edge("Ch",i,color=Colors["Ch"])
for i in CO:
  g.add_edge("Co",i,color=Colors["Co"])

##find "None"
h=dict(g.nodes(data="group"))
k,v=list(h.keys()),list(h.values())

for i in range(8,len(k)):
  if v[i] is None:
    g.nodes[k[i]]["group"]="alone"

pos = nx.circular_layout(g)
h=net.Network(notebook=True)
#h.toggle_physics(False)
h.repulsion(node_distance=120, central_gravity=0.0, spring_length=100, spring_strength=0.01, damping=0.09)
h.from_nx(g)
h.show_buttons(filter_=["physics"])
h.show("example.html")
```

```python
g = Graph('G',filename='cluster.gv',engine="twopi")
g.attr(start="Regular",ranksep="10",overlap="prism",ratio="auto")
g.edge_attr.update(color="lightgrey")

g.node('A', 'AMEX MR',color='yellow')
g.node("B","Bilt",color='honeydew3')
g.node('Ci', 'Citi',color='red')
g.node('Ch', 'Chase',color='skyblue1')
g.node("Co","Capital One",color='green')
g.node("c","cards",root="true",color="white")
g.edges(["cA","cB",("c","Ci"),("c","Ch"),("c","Co")])

with g.subgraph(name='cluster_star') as c:
    c.attr(color='lightgrey',overlap="prism", rank="max")
    c.edge_attr.update(color="lightgrey")
    for i in range(len(star)):
      j_l=list(range(i+1,len(star)))
      c.edges([(star[i],star[j]) for j in j_l])
      #c.node(star[i])
    c.attr(label='Star Alliance')
    c.node_attr.update(style='filled', color='lightgray',overlap="prism")


with g.subgraph(name='cluster_one') as o:
    o.attr( color='lightgrey',overlap="false", rank="max")
    o.edge_attr.update(color="lightgrey")
    for i in range(len(one)):
      j_l=list(range(i+1,len(one)))
      o.edges([(one[i],one[j]) for j in j_l])
      #o.node(one[i])
    o.attr(label='Oneworld')
    o.node_attr.update(style='filled', color='plum1',overlap="prism")

with g.subgraph(name='cluster_sky') as s:
    s.attr( color='lightgrey',overlap="false", rank="max")
    for i in range(len(sky)):
      j_l=list(range(i+1,len(sky)))
      s.edges([(sky[i],sky[j]) for j in j_l])
      #s.node(sky[i])
    s.attr(label='SkyTeam')
    s.node_attr.update(style='filled', color='lightblue',overlap="prism")

AS=["Aer Lingus","Air Tahiti Nui", "Condor","Israel","Hainan","Icelandair","Korean Air","LATAM","Ravn Alaska","Singapore","STARLUX"]
for i in AS:
  g.edge("Alaska",i)
VA=[]

for i in Bilt:
  g.edge("B",i,color='honeydew3')
for i in AMEX:
  g.edge("A",i,color="yellow")
for i in Citi:
  g.edge("Ci",i,color="red")

for i in Chase:
  g.edge("Ch",i,color='blue')
for i in CO:
  g.edge("Co",i,color="green")

g.node_attr.update(style='filled', color="white",overlap="prism")

g.attr(beautify="true",overlap="prism")
```
